```python
import cv2
import matplotlib.pyplot as plt
cvt=cv2.imread('img.jpg',0)
plt.imshow(cvt)
plt.show()
```

```python
hist=cv2.calcHist([cvt],[0],None,[256],[0,256])
plt.figure()
plt.title("Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixel count")
plt.stem(hist)
plt.show()
```

```python
equ=cv2.equalizeHist(cvt)
```
